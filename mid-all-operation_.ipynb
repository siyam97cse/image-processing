```python
import matplotlib.pyplot as plt
import os
import pandas as pd
import numpy as np
```

```python
import cv2
```

```python
from skimage import exposure
```

```python
img_path='C:/Users/Student/Downloads/lionn.jpg'
```

```python
img=cv2.imread(img_path)
```

```python
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.imshow(img1)
plt.axis("off")
plt.title('Original')

plt.subplot(1,2,2)
plt.imshow(grey)
plt.axis("off")
plt.title('Grey')
```

```python
b=1.2
b_img=cv2.convertScaleAbs(img1,alpha=b,beta=0)
```

```python
c=2.1
c_img=cv2.addWeighted(img1,c,np.zeros_like(img1),0,0)
```

```python
plt.figure(figsize=(16,10))
plt.subplot(1,3,1)
plt.imshow(b_img)
plt.axis("off")
plt.title('Brighted')

plt.subplot(1,3,2)
plt.imshow(c_img)
plt.axis("off")
plt.title('contrast')
```

```python
g_img=exposure.adjust_gamma(img1,gamma=1,gain=1)
```

```python
l_img=exposure.adjust_log(img1,gain=1,inv=False)
```

```python
s_img=exposure.adjust_sigmoid(img1,cutoff=0.5,gain=10,inv=False)
```

```python
plt.figure(figsize=(16,10))
plt.subplot(1,3,1)
plt.imshow(g_img)
plt.axis("off")
plt.title('gamma')

plt.subplot(1,3,2)
plt.imshow(l_img)
plt.axis("off")
plt.title('log')

plt.subplot(1,3,3)
plt.imshow(s_img)
plt.axis("off")
plt.title('sigmoid')
```

```python
c_img=exposure.cumulative_distribution(img1,nbins=256)
```

```python
e_img=exposure.equalize_adapthist(img1,kernel_size=None,clip_limit=0.01,nbins=256)
```

```python
plt.figure(figsize=(16,10))
plt.subplot(1,2,1)
plt.imshow(c_img)
plt.axis("off")
plt.title('cumulative')

plt.subplot(1,2,2)
plt.imshow(e_img)
plt.axis("off")
plt.title('equalizr')
```

```python
img2=cv2.imread('C:/Users/Student/Downloads/man1.jpg')
img3=cv2.imread('C:/Users/Student/Downloads/man2.jpg')
```

```python
re_img1=cv2.resize(img2,(400,400))
re_img2=cv2.resize(img3,(400,400))
```

```python
add_img=cv2.add(re_img1,re_img2)
```

```python
mul_img=cv2.multiply(re_img1,re_img2)
```

```python
sub_img=cv2.subtract(re_img1,re_img2)
```

```python
div_img=cv2.divide(re_img1,re_img2)
```

```python
plt.figure(figsize=(16,10))
plt.subplot(1,4,1)
plt.imshow(add_img)
plt.axis("off")
plt.title('Addition')

plt.subplot(1,4,2)
plt.imshow(mul_img)
plt.axis("off")
plt.title('multiply')

plt.subplot(1,4,3)
plt.imshow(sub_img)
plt.axis("off")
plt.title('Subtract')

plt.subplot(1,4,4)
plt.imshow(div_img)
plt.axis("off")
plt.title('divide')
```
